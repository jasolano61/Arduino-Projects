Existen muchos ceros visibles.
Establecer cuántos valores son cero en todas las columnas de fechas:

```python
tot_valores = (df.iloc[:, 4:] >= 0).sum().sum()
Val_en_Cero = (df.iloc[:, 4:] == 0).sum().sum()
Porc_en_cero = (Val_en_Cero / tot_valores) * 100
print(f"Total de valores: {tot_valores}")
print(f"Total de valores en cero: {Val_en_Cero}")
print(f"Porcentaje de valores en cero: {Porc_en_cero:.2f}%")
```

Evolucion de los casos en USA

```python
import matplotlib.pyplot as plt

pais = "Greece"
datos_pais = df[df["Country/Region"] == pais].iloc[:, 4:].sum()

print (datos_pais.count())

plt.figure(figsize=(12, 5))
plt.plot(datos_pais.index, datos_pais.values)
plt.xticks(rotation=45)
plt.title(f"Evolución de casos en {pais}")
plt.xlabel("Fecha")
plt.ylabel("Casos acumulados")
plt.show()
```
